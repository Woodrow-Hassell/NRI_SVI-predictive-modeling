```python
import numpy as np
import pandas as pd
```

```python
NRI_df = pd.read_csv('NRI Mecklenburg County Census Tracts.csv')
SVI_df = pd.read_csv('SoVI Mecklenburg County Census Tracts.csv')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

NRI_df.head()
```

```python
#Identifying columns with the same values in every row
counts = NRI_df.nunique()
to_del = [i for i, v in enumerate(counts) if v == 1]
print(to_del)

#Identifying columns with no values
counts2 = NRI_df.nunique()
to_del2 = [i for i, v in enumerate(counts) if v == 0]
print(to_del2)
```

```python
# Combine the lists
columns_to_drop = list(set(to_del + to_del2))

# Drop columns from the DataFrame
NRI_df.drop(columns=NRI_df.columns[columns_to_drop], inplace=True)

#print(NRI_df.nunique())
```

```python
#Dropping columns of hazards not in use
prefixes_to_drop = ['AVLN', 'HAIL', 'CFLD', 'CWAV', 'DRGT', 'ERQK', 'HRCN', 'LNDS', 'TRND', 'TSUN', 'VLCN', 'WFIR']
columns_to_drop = [col for col in NRI_df.columns if any(col.startswith(prefix) for prefix in prefixes_to_drop)]
NRI_df.drop(columns=columns_to_drop, inplace=True)
```

```python
# Select columns with string data type
string_columns = NRI_df.select_dtypes(include=['object']).columns

# Exclude specific columns
columns_to_exclude = ['NRI_ID', 'GlobalID']
filtered_columns = [col for col in string_columns if col not in columns_to_exclude]

# Loop through each column and print unique values
for column in filtered_columns:
    unique_values = NRI_df[column].unique()
    print(f"Column: {column}")
    print(f"Values: {unique_values}\n")
```

```python
columns_of_interest = ['RISK_RATNG', 'EAL_RATNG', 'SOVI_RATNG', 'HWAV_HLRR',
                        'HWAV_EALR', 'HWAV_RISKR', 'ISTM_HLRR', 'ISTM_EALR',
                        'ISTM_RISKR', 'LTNG_HLRR', 'LTNG_EALR', 'LTNG_RISKR',
                        'RFLD_HLRR', 'RFLD_EALR', 'RFLD_RISKR', 'SWND_HLRR',
                        'SWND_EALR', 'SWND_RISKR', 'WNTW_HLRR', 'WNTW_EALR', 'WNTW_RISKR']

# Get unique values across all specified columns
all_unique_values = np.unique(NRI_df[columns_of_interest].values.flatten())

# Print the unique values
print(all_unique_values)
```

```python
# Mapping risk value strings to numeric

value_mapping = {
    'No Expected Annual Losses': 0,
    'No Rating': 0,
    'Very Low': 1,
    'Relatively Low': 2,
    'Relatively Moderate': 3,
    'Relatively High': 4,
    'Very High': 5
}

# Apply the mapping to the specified columns
NRI_df[columns_of_interest] = NRI_df[columns_of_interest].replace(value_mapping)
```

```python
#Checking Variance in each columns compared to total number of rows

for i in range(NRI_df.shape[1]):
    num = NRI_df.iloc[:, i].nunique()
    perc = float(num) / NRI_df.shape[0] * 100
    if perc < 1:
        print(f"{NRI_df.columns[i]}: {num} unique values, {perc:.1f}%")
```
